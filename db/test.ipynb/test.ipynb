```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.sisul.or.kr/open_content/traffic/bbs/bbsMsgList.do?bcd=faq&listsz=10&symd=&eymd=&cate1=07&keyfield=title&keyword="
response = requests.get(url)
bs = BeautifulSoup(response.text, "html.parser")
print(url)
```

```python
q= bs.select_one('thead tr')
print(q.text)

a= bs.select_one('tbody p')
print(a.text)

b= bs.select_one('tr td')
print(b.text)
```
